```python
import csv
import pandas as pd
import numpy as np
import datetime as dt
import xlsxwriter
```

```python
df1 = pd.read_excel('Data.xlsx',sheet_name= 'Sales Data')
df2 = pd.read_excel('Data.xlsx',sheet_name= 'Master Data')
df1.rename(columns = {'Store Code':'Store'}, inplace = True)
```

```python

df2.drop_duplicates(inplace=True)
df = pd.merge(df1,df2,how="inner",on=["Store"])
```

```python
df['Payment']= list(map(lambda x,y: 599 if((x=='Product A') & (y==1)) else(799 if((x=='Product B') & (y==1)) else(-599 if((x=='Product A') & (y==-1)) else(-799 if((x=='Product B') & (y==-1)) else(0)))) , df['Product Name'],df['Sales Volume'] ))

```

```python

df['Staff_Incentive']= list(map(lambda x,y: 35  if((x=='Product A') & (y==1)) else(70 if((x=='Product B') & (y==1)) else(0)) , df['Product Name'],df['Sales Volume']))
```

```python
# GENERAL MANAGER
GM= df.groupby("General Manager").agg({"Sales Volume":np.sum , "Store":pd.Series.nunique})
GM['ASPS']= GM['Sales Volume']/GM['Store']
GM['Incentive'] = list(map(lambda x,y:y*(6.66) if x>125 else( y*(133.33) if(x>150) else(0)),GM['ASPS'],GM['Sales Volume']))
GM['Actual_Incentive']= list(map(lambda x: x if(x<=120000) else(120000),GM['Incentive']))
GM
```

```python
# Area Manager
AM = df.groupby("Area Manager").agg({"Sales Volume":np.sum , "Store":pd.Series.nunique})
AM['ASPS']= AM['Sales Volume']/AM['Store']
AM['Incentive'] = list(map(lambda x,y:y*(25) if x>100 else( y*(50) if(x>150) else(0)),AM['ASPS'],AM['Sales Volume']))
AM['Actual_Incentive']= list(map(lambda x: x if(x<=100000) else(100000),AM['Incentive']))
AM
```

```python
# Staff Incentive
SI = df.groupby("Store").agg({"Staff_Incentive":np.sum})
SI
```

```python
# Store Manager
SM = df.groupby(['Store','Product Name']).agg({"Sales Volume":np.sum})

SM .reset_index(inplace=True)
SM_Pivot = SM.pivot(index= 'Store' ,columns='Product Name', values =['Sales Volume'])

SM_Pivot['%age Product B'] = (SM_Pivot['Sales Volume']['Product B']/(SM_Pivot['Sales Volume']['Product B']+ SM_Pivot['Sales Volume']['Product A']))*100
SM_Pivot['Incentive'] = list(map(lambda x,y:(36.6)*y if x>=50 else( y*(33.3) if(x<50) else(0)),SM_Pivot['%age Product B'],SM_Pivot['Sales Volume']['Product B']))
SM_Pivot


```
